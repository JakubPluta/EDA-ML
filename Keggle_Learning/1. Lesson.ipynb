```python
# Line plots
f,ax = plt.subplots(1,1,figsize=(15,8))
a2 = data['Speed'].plot(kind='line',color='#0B486B',label='Speed',linewidth=1.1,ax=ax,grid = True,linestyle = ':')
a3 = data['Defense'].plot(kind='line',color='#c55572',label='Defense',linewidth=1.1,ax=ax,grid = True,linestyle = '-.')
plt.legend(loc='upper left');
plt.xlabel('x axis')
plt.ylabel('y axis');
plt.title('Defense&Speed Line Plots',fontsize=15)
```
