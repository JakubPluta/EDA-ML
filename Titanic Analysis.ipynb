```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.head()
```

```python
train.drop(train[pd.isnull(train['Embarked'])].index, inplace=True)
```

```python
train.sample(3)
```

```python
train.isnull().sum()
```

```python
def age_group(age):
    if age < 18:
        return '<18'
    elif age >=18 and age <30:
        return '18-30'
    elif age >=30 and age <50:
        return '30-50'
    else:
        return '>50'
```

```python
train['Age_Group'] = train['Age'].apply(lambda x: age_group(x))
```

```python
f,ax = plt.subplots(2,2,figsize=(16,10))
x = train.groupby('Age')['Age'].count()
y = test.groupby('Age')['Age'].count()
a1 = sns.distplot(x,label='train',ax=ax[0][0])
a2 = sns.distplot(y,label='test',ax=ax[0][0])
a1.legend(loc='best')
a1.set_title("Age distribution");

b1= train.Sex.value_counts().plot(kind='barh', color='#556270', label='train',ax=ax[0][1])
b2 = test.Sex.value_counts().plot(kind='barh', label='test', color='#C44D58',ax=ax[0][1])
b1.set_ylabel('Sex')
b1.set_title("Sex distribution" )
b1.legend(loc='best');

plt.suptitle('Distribution of some features of titanic dataset train/test',fontsize=15)
c1= train.Embarked.value_counts().plot(kind='barh', color='#11999e', label='train',ax=ax[1][0])
c2 = test.Embarked.value_counts().plot(kind='barh', label='test', color='#311d3f',ax=ax[1][0])
c1.set_ylabel('Embarked')
c1.set_title("Embarked distribution" )
c1.legend(loc='best');


d1= train.Embarked.value_counts().plot(kind='barh', color='#311d3f', label='train',ax=ax[1][1])
d2 = test.Embarked.value_counts().plot(kind='barh', label='test', color='#88304e',ax=ax[1][1])
d1.set_ylabel('Pclass')
d1.set_title("Pclass distribution" )
d1.legend(loc='best');


plt.figure(figsize=(16,4))
n = train.groupby('Pclass')['Pclass'].count()
m = test.groupby('Pclass')['Pclass'].count()
v1 = sns.distplot(n,label='train')
v2 = sns.distplot(m,label='test')
v1.legend(loc='best')
v1.set_title("Pclass distribution");
```

```python
features = ['Pclass','Sex','SibSp','Parch','Embarked','Age_Group']
f,ax = plt.subplots(2,3,figsize=(12,8))
start = 0
for j in range(2):
    for i in range(3):
        if start == len(features):
            break
        sns.barplot(x=features[start],y='Survived',data=train,ax=ax[j, i],palette='twilight_r')
        start +=1
```

After short analysis we can see few insights:

* Survival chances of women are much higher then man.
* First Pclass has the highest Survival chances. Better class > more chances to survive.
* People who was travelling alone or has a lot of sibblings has lower survival chances then people who was travelling with 1-2 sibblings.
* Embarked C has highest survival rate. 
* People who are under 18 has most chances to survive.
