## Zbiór danych gapminder

Analizowane dane dotyczą wskaźników ekonomicznych w krajch z całego świata na przełomie lat 1952-2007.

Znaczenie kolumn:


* country - kraj z którego pochodzą dane,
* year - rok badania, 
* pop - populacja danego kraju,
* continent - kontynent, na którym leży dany kraj,
* lifeExp - średnia długość życia,
* gdpPercap - GDP (produkt krajowy brutto) w przeliczeniu na 1 mieszkańca.

**Pierwszy krok ** to zaimportowanie biblioteki pandas (dostępnej w Google Colab) oraz wczytanie danych do zmiennej df. Kod wykonujący te czynności znajduje się w dwóch poniższych komórkach, aby je uruchomić wciśnij Shift+Enter.


```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv')
```

**W jakich krajach średnia długośc życia była najkrótsza w roku 1952?**

```python
df[df['year']==1952].sort_values(by='lifeExp').head(5)
```

**W jakich krajach średnia długość życia była najdłuższa w roku 2007?**

**W jakich krajach GDP per capita było najwyższe w roku 2007 i najniższe w roku 1952?**

Najniższe

```python
gdp_1952_asc = df[df['year']==1952].sort_values(by='gdpPercap',ascending=True)
gdp_1952_asc.head(10)
```

Najwyższe

```python
gdp_2007_desc = df[df['year']==2007].sort_values(by='gdpPercap',ascending=False)
gdp_2007_desc.head(10)
```

**Jaka jest korelacja między GDP, a średnią długością życia?**

```python
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = "{:.2f}".format
```

```python
df.describe()[['lifeExp','gdpPercap']]
```

```python
f,ax = plt.subplots(figsize=(12,4)) # Sporo obserwacji odstających przy gdpPerCapita
sns.scatterplot(x="lifeExp", y="gdpPercap",           
                palette='inferno_r',hue='continent',  linewidth=0, alpha=0.9,
                data=df);
plt.title(f'Korelacja pomiędzy gpdPercap & lifeExp',fontsize=14);
# Korelacja istnieje, wraz z wzrostem gdpPerCap wzrasta lifExp 
```

```python
f,ax = plt.subplots(figsize=(12,4)) # Sporo obserwacji odstających przy gdpPerCapita
sns.scatterplot(x="lifeExp", y="gdpPercap",           
                palette='plasma_r',hue='continent',  linewidth=0, alpha=0.9,
                data=df[df['gdpPercap']<=9300]); ## df['gdpPercap'].mean()
plt.title(f'Korelacja pomiędzy gpdPercap & lifeExp gdzie gpdPerCap < 9300 kwartyl Q3',fontsize=14);
# Korelacja istnieje, wraz z wzrostem gdpPerCap wzrasta lifExp 
```

```python
corr = df['gdpPercap'].corr(df['lifeExp'])
corr
```

```python
print(f'Korelacja pomiędzy gdpPercap i lifeExp {corr}. Dosyć silna korelacja dodatnia.')
```

**Gdzie warunki życia poprawiły się najbardziej?**
Procentowo według średniej długości życia porównując rok 1952 i 2007



```python
df_piv = df[(df['year']==1952) | (df['year']==2007)]
df_piv = df_piv.pivot_table(values='lifeExp',index='country',columns='year',aggfunc='mean')
df_piv = df_piv.reset_index()
df_piv['pct_change'] = (df_piv[2007]/df_piv[1952]-1)*100
df_piv = df_piv.sort_values(by='pct_change',ascending=False)
df_piv = df_piv[['country',1952,2007,'pct_change']].head(10).reset_index(drop=True)
df_piv
```

```python
# Wskazówka: pomocne będzie zagregowanie danych funkcją pivot_table.
```

## Wykresy i wizualizacje

```python
# importowanie biblioteki do wizualicaji danych
import matplotlib.pyplot as plt
import seaborn as sns #Użyje Seaborn
```

**Jak średnia długość życia zależy od GDP per capita (rok 2007)?**

```python
d2007 = df[df['year']==2007]
sns.set(style='darkgrid')
f,ax = plt.subplots(figsize=(8,6)) 
sns.scatterplot(y="lifeExp", x="gdpPercap",           
                color='b',  linewidth=0, alpha=1,s=90,
                data=d2007);
plt.title(f'Korelacja pomiędzy gpdPercap & lifeExp 2007',fontsize=14);
```

```python
f,ax = plt.subplots(figsize=(8,6)) 
sns.scatterplot(y="lifeExp", x="gdpPercap",           
                color='b',  linewidth=0, alpha=1,palette='plasma_r',hue='continent',s=90,
                data=d2007);
plt.title(f'Korelacja pomiędzy gpdPercap & lifeExp 2007 w podziale na kontynent',fontsize=14);
```

### Statystyki - Polska

**Średnia długość życia w Polsce**

```python
f,ax = plt.subplots(figsize=(12,6))
df_pl = df[df['country']=='Poland']
sns.lineplot(y = df_pl['lifeExp'],x=df_pl['year'],color='gray',linewidth=4)
plt.title(f'Average lifeExp in Poland',fontsize=14);
```

### Statystyki - świat

**Zmiana długości życia per kontynent**

```python
continents = (df.groupby(['continent','year']).mean()).reset_index()
continents.head(5)
```

```python
f,ax = plt.subplots(figsize=(12,8))
sns.lineplot(y='lifeExp',x='year',data= continents,hue='continent',palette='cividis',linewidth=5,style='continent')
plt.legend(loc='upper left');
plt.title('Zmiana średniej długości życia w czasie w podziale na kontynent',fontsize=14);
```

**Zmiany w średniej wartości GDP per capita w zależności od kontynentu**

```python
#continents['gdpPercap'].head(5)
continents.head(5)
```

```python
f,ax = plt.subplots(figsize=(12,8))
sns.lineplot(y='gdpPercap',x='year',data= continents,hue='continent',palette='cubehelix',linewidth=5)
plt.legend(loc='upper left');
plt.title('Zmiana GDP w czasie w podziale na kontynenty',fontsize=14);
```

W tej części czekamy na Twoje pomysły odnośnie analizy tego zbioru danych. Opisz dokładnie co chcesz sprawdzić i na jakie pytania szukasz odpowiedzi.

```python
#country - kraj z którego pochodzą dane,
#year - rok badania,
#pop - populacja danego kraju,
#continent - kontynent, na którym leży dany kraj,
#lifeExp - średnia długość życia,
#gdpPercap - GDP (produkt krajowy brutto) w przeliczeniu na 1 mieszkańca.
```

```python
f,ax = plt.subplots(figsize=(8,6))
sns.heatmap(df.corr(),cmap='YlGnBu',annot=True,linewidths=.1,square=True);
plt.title('Macierz korelacji',fontsize=14);
```

Widać, że nie znacznej korelacji pomiędzy pozostałymi zmiennymi. Dlatego dalsza analiza zależności pomiędzy zmiennymi nie ma sensu. Móżna zbadać jak zmieniały się różne zjawiska w czasie w podziale na kraj/kontynent. 

```python
popu = (df.groupby(['continent','year'])['pop'].sum()).reset_index()
popu['pct_change'] = popu['pop'].pct_change()
popu.head()
```

```python
# Średnie tempo wzrostu populacji w podziale na kontynenty
(popu.groupby('continent')['pct_change'].mean()).sort_values(ascending=False)
```

```python
f,ax = plt.subplots(figsize=(14,12))
sns.lineplot(y='pop',x='year',data=popu,hue='continent' ,palette='cubehelix',linewidth=5)
plt.ticklabel_format(style='plain', axis='y')
plt.legend(loc='upper left');
plt.title('Populacja w czasie w podziale na kontynenty',fontsize=14);
```
