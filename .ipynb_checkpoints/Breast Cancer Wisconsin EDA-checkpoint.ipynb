```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings("ignore")
from scipy import stats
import random
from matplotlib import cm
from _plotly_future_ import v4_subplots
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
#from _plotly_future_ import v4_subplots
```

```python
data = pd.read_csv(r'Breast Cancer\bc.csv')
data.head()
```

```python
data.columns
```

```python
def random_colors(number_of_colors):
    color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
                 for i in range(number_of_colors)]
    return color


def dist(column, colors = ['#2F4858','#F6AE2D'],data=data,bins=False ) :  
    dfB,dfM = data[data['diagnosis']=='B'][column], data[data['diagnosis']=='M'][column]
    hist = [dfB,dfM]
    labels = ['benign','malignant']
    colors = colors
    f = ff.create_distplot(hist, labels, colors = colors, show_hist = True, curve_type='kde',bin_size=bins)
    f['layout'].update(title = column,autosize=False,
    width=800,
    height=300)
    py.iplot(f, filename = 'Density plot')

    
    
def annot(chart):
    for p in chart.patches:
        chart.annotate("%.2f" % p.get_height(),
                       (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```

```python
data_wo = data.drop(['Unnamed: 32','id','diagnosis'],axis=1)
data_d = data.diagnosis
```

```python
f,ax = plt.subplots(1,2,figsize=(16,6))
q1 = sns.barplot(x=['Benign','Malignant'],
                 y=data['diagnosis'].value_counts().values,palette=['#2184A0','#9B89B3'],ax=ax[0])
sns.despine(left=True,bottom=True);
annot(q1)
c1 = cm.cividis(np.linspace(.1,1.06, 2))
my_circle=plt.Circle( (0,0), 0.83, color='white')
circle1 = plt.pie(data['diagnosis'].value_counts().values, labels = ['Benign','Malignant'], 
        colors = ['#2184A0','#9B89B3'], 
        autopct = '%1.1f%%');
p=plt.gcf()
p.gca().add_artist(my_circle);
plt.setp(plt.gcf().get_axes(), xlabel='', ylabel='');
plt.suptitle('Benign & Malignant Breast Cancer Wisconsin',fontsize=15);
```

```python
normalized = (data_wo - data_wo.mean()) / (data_wo.std())  
```

```python
data_normalized = pd.concat([data_d,normalized],axis=1)
```

```python
melted1 = pd.melt(data_normalized,id_vars="diagnosis",
                    var_name="columns",
                    value_name='values')
```

```python
melted1a =  pd.melt(pd.concat([data_d,normalized.iloc[:,0:10]],axis=1),id_vars="diagnosis",
                    var_name="columns",
                    value_name='values')
melted1b =  pd.melt(pd.concat([data_d,normalized.iloc[:,10:20]],axis=1),id_vars="diagnosis",
                    var_name="columns",
                    value_name='values')

melted1c =  pd.melt(pd.concat([data_d,normalized.iloc[:,20:30]],axis=1),id_vars="diagnosis",
                    var_name="columns",
                    value_name='values')
```

```python
plt.figure(figsize=(16,8))
sns.violinplot(x="columns", y="values", hue="diagnosis",inner='quart', data=melted1,split=True,palette=['#2184A0','#9B89B3'])
plt.xticks(rotation=90);
plt.xlabel('');
plt.ylabel('');
sns.despine(left=True,bottom=True);
```

```python
plt.figure(figsize=(16,4))
sns.boxenplot(x="columns", y="values", hue="diagnosis", data=melted1a,palette=['#2184A0','#9B89B3'])
plt.xlabel('');
plt.ylabel('');
sns.despine(left=True,bottom=True);
plt.figure(figsize=(16,4))
sns.boxenplot(x="columns", y="values", hue="diagnosis", data=melted1b,palette=['#33658A','#F26419'])
plt.xlabel('');
plt.ylabel('');
sns.despine(left=True,bottom=True);
plt.figure(figsize=(16,4))
sns.boxenplot(x="columns", y="values", hue="diagnosis", data=melted1c,palette=['#2F4858','#F6AE2D'])
plt.xlabel('');
plt.ylabel('');
sns.despine(left=True,bottom=True);
```
