```python
cols = ['country','year','sex','age','SuicidesPer100k']
train_1 = train[cols]
train_1.head()
```

```python
def decades(year):
    if year >= 1987 and year <= 1996:
        return '1987-1996'
    elif year >= 1997 and year <= 2006:
        return '1997-2006'
    else:
        return '2007-2016'
```

```python
train_1['decade'] = train_1['year'].apply(decades)
train_1.sample(5)
```

```python
train_1 = train_1[['country','sex','age','decade','SuicidesPer100k']]
```

```python
danger = train_1.groupby(["country","decade"]).sum().reset_index()
danger.head()
```

```python
three_gen = danger.country.value_counts().reset_index()
three_gen.columns = ["country", "counts"]
three_gen_countries = three_gen[three_gen.counts == 3].country.tolist()
#len(three_gen_countries)
```

```python
decades_3 = danger.country.value_counts().reset_index()
decades_3 = decades_3[decades_3['country']==3]['index'].tolist()
dec = danger['decade'].unique().tolist()
green_zones, danger_zones = [],[]
```

```python
for country in decades_3:
    s_year1 = float(danger[(danger['country']==country) & (danger['decade']=='1987-1996')]['SuicidesPer100k'])
    s_year2 = float(danger[(danger['country']==country) & (danger['decade']=='1997-2006')]['SuicidesPer100k'])
    s_year3 = float(danger[(danger['country']==country) & (danger['decade']=='2007-2016')]['SuicidesPer100k'])
    if s_year1 <= s_year2 <= s_year3:
        danger_zones.append(country)
    if s_year1 >= s_year2 >= s_year3:
        green_zones.append(country)
                                                                                     
```

```python
decrease = danger[danger['country'].isin(green_zones)]
increase = danger[danger['country'].isin(danger_zones)]
plt.figure(figsize=(16,8))
sns.barplot(x = "country", y = "SuicidesPer100k", hue = "decade",data = decrease,palette='cividis_r').set_title("Decreasing Suicide Rate",fontsize=16)
plt.xticks(rotation = 90);
```

These are countries where Suicide Rate decreased by decades

```python
plt.figure(figsize=(16,8))
sns.barplot(x = "country", y = "SuicidesPer100k", hue = "decade",data = increase,palette='cividis_r').set_title("Increase Suicide Rate",fontsize=16)
plt.xticks(rotation = 90);
```

These are countries where Suicide Rate increased by decades
