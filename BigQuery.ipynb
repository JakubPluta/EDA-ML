```python
%%bigquery
SELECT U.display_name, U.age, H.*
FROM `bigquery-public-data.stackoverflow.users` AS U
LEFT JOIN `bigquery-public-data.stackoverflow.post_history` AS H ON U.id = H.user_id
LIMIT 3
```

```python
%%bigquery
SELECT
EXTRACT(YEAR FROM creation_date) AS Year,
COUNT(*) AS Number_of_Questions,
SUM(answer_count) AS Answered_Questions,
SUM(comment_count) AS Commented_Questions,
SUM(view_count) AS Viewed_Questions
FROM
  `bigquery-public-data.stackoverflow.posts_questions`
GROUP BY
  Year
HAVING
  Year > 2008 AND Year <= 2018
ORDER BY Year; 
```
