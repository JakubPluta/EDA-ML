```python
sqlA = '''
  SELECT
        EXTRACT(year from questions.creation_date) as Year,
        COUNT(distinct questions.id) as Number_of_Questions,
        COUNT(*) as Number_of_Answers
  FROM 
            `bigquery-public-data.stackoverflow.posts_questions` questions
  LEFT JOIN 
            `bigquery-public-data.stackoverflow.posts_answers` answers
  ON 
            questions.id = answers.parent_id
  WHERE 
            questions.tags like "%python%"
  GROUP BY 
          Year
  ORDER BY 
          1 ASC

'''

dfA = client.query(sqlA).to_dataframe()
dfA['Ratio'] = dfA['Number_of_Answers']/dfA['Number_of_Questions']
```

```python
plt.figure(figsize=(16,4))
sns.lineplot(data=dfA.set_index('Year')[['Number_of_Questions','Number_of_Answers']],palette=['#FE938C','#3A506B'],linewidth=5).set_title('Number of Answers vs Number of Questions ',fontsize=22)
sns.despine(top=True, right=True, left=True, bottom=True, offset=None, trim=True)
plt.xlabel('')
f,ax = plt.subplots(1,2,figsize=(15,5),sharey='row')
a1 = sns.barplot(y='Number_of_Questions',x='Year',data=dfA,palette=['#FE938C'],ax=ax[0])
a1.set(ylim=(0, 350000))
plt.xlabel('')
a2 = sns.barplot(y='Number_of_Answers',x='Year',data=dfA,palette=['#3A506B'],ax=ax[1])
sns.despine(top=True, right=True, left=True, bottom=True, offset=None, trim=True)
plt.xlabel('')
plt.figure(figsize=(16,4))
sns.lineplot(data=dfA.set_index('Year')['Ratio'],color='#79BD9A',linewidth=7).set_title('Ratio Answers/Questions')
sns.despine(top=True, right=True, left=True, bottom=True, offset=None, trim=True);
dfA
```

```python
sql0 = '''

SELECT
EXTRACT(YEAR FROM creation_date) AS Year,
SUM(comment_count) AS Number_of_Comments,
SUM(view_count) AS Number_of_views,
COUNT(distinct id) Questions
FROM
  `bigquery-public-data.stackoverflow.posts_questions` WHERE tags like '%python%'
GROUP BY
  Year
HAVING
  Year >= 2008 AND Year <= 2019
ORDER BY Year; 

'''

df0 = client.query(sql0).to_dataframe()
df0['Comments Ratio'] = df0['Number_of_Comments']/df0['Questions']
df0['Views Ratio'] = df0['Number_of_views']/df0['Questions']
```

```python
f,ax = plt.subplots(2,1,figsize=(15,8),sharex='col')
a3 = sns.lineplot(y=df0['Comments Ratio'],x=df0['Year'],data=df0,color='#FE938C',ax=ax[0],linewidth=5)
plt.xlabel('');
a4 = sns.lineplot(y=df0['Views Ratio'],x=df0['Year'],data=df0,color='#3A506B',ax=ax[1],linewidth=5)
sns.despine(top=True, right=True, left=True, bottom=True, offset=None, trim=False)
plt.suptitle('Comments/Questions Ratio & Views/Questions Ratio',fontsize=22);
df0
```

```python
sql2 = """
SELECT 
  EXTRACT(YEAR FROM creation_date) year, 
  IF(title LIKE '%?', 'with?', 'without') ends_with_question,
  ROUND(COUNT(accepted_answer_id )* 100/COUNT(*), 2) as answered 
FROM `bigquery-public-data.stackoverflow.posts_questions`
WHERE tags like '%python%'
GROUP BY 1,2
ORDER BY 1,2
"""



df2 = client.query(sql2).to_dataframe()
```

```python
df2
```

```python
plt.figure(figsize=(16,6))
sns.barplot(y='answered',x='year',data=df2,palette='cividis',hue='ends_with_question')
sns.despine(top=True, right=True, left=True, bottom=True, offset=None, trim=False)
plt.title('% of Answered Questions with Quote mark in title or without quote mark in title',fontsize=22);
```

```python
sql3 = """SELECT 
  EXTRACT(YEAR FROM creation_date) year, 
  IF(title LIKE '%?', 'with?', 'without') ends_with_question,
  COUNT(*) as cnt
FROM `bigquery-public-data.stackoverflow.posts_questions`
WHERE tags like '%python%'
GROUP BY 1,2
ORDER BY 1,2
"""



df3 = client.query(sql3).to_dataframe()
df3
```

```python
plt.figure(figsize=(16,8))
sns.barplot(y='cnt',x='year',data=df3,palette='cividis',hue='ends_with_question')
sns.despine(top=True, right=True, left=True, bottom=True, offset=None, trim=False)
plt.title('Questions with quote mark  in title, per year',fontsize=22);
```

```python
sql4 = '''
    select 
    EXTRACT(YEAR FROM creation_date) year,
    count(distinct id) as questions, count(accepted_answer_id) as accepted_answers
    from `bigquery-public-data.stackoverflow.posts_questions`
    where 
    tags like '%python%'
    group by 1 order by 1 asc
'''
df4 = client.query(sql4).to_dataframe()
df4.set_index('year',inplace=True)
```

```python
tags = '''
    select id, tags,EXTRACT(YEAR FROM creation_date) year
        from `bigquery-public-data.stackoverflow.posts_questions`
            where
            tags like '%python%'
'''
df5 = client.query(tags).to_dataframe()

df5.head()
```

```python
rows_list = []
for _, rows in df5.iterrows():
    tag = rows.tags.split('|')
    for t in tag:
        if t != 'python':
            row = {'question_id': rows.id, 'tag': t}
            rows_list.append(row)
tags_per_question = pd.DataFrame(rows_list)
tag_count = tags_per_question.groupby('tag').count().sort_values(by='question_id', ascending=False).rename(columns={'question_id':'counts'})
tag_count.head(30)
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x=tag_count.index[0:30], y=tag_count.counts[0:30],palette=['#3A506B']);
plt.xticks(rotation=90);
sns.despine(top=True, right=True, left=True, bottom=True, offset=None, trim=False)
plt.title('Most popular tags with Python', fontsize=22);
```

```python
df5['tags_count'] = df5['tags'].str.split('><').str.len()
```

```python
tags_year = pd.DataFrame(df5.groupby(['year','tags_count'])['tags_count'].count()).rename(columns={'tags_count':'counts'}).reset_index()
tags_year.head()
```

```python
plt.figure(figsize=(16,4))
g = sns.barplot(x=tags_year['year'], y=tags_year['counts'],palette='tab20b_r',hue=tags_year['tags_count'])
plt.xticks(rotation=90);
sns.despine(top=True, right=True, left=True, bottom=True, offset=None, trim=False)
plt.suptitle('Number of tags with Pythona Questions', fontsize=18);
plt.figure(figsize=(16,2))
tags_all = pd.DataFrame(df5.groupby(['tags_count'])['tags_count'].count()).rename(columns={'tags_count':'counts'}).reset_index()
sns.barplot(x="counts", y="tags_count",
                data=tags_all, palette="cividis",orient = 'h');
sns.despine(top=True, right=True, left=True, bottom=True, offset=None, trim=False)
```
