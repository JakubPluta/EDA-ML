```python
#plt.figure(figsize=(12,6))
f,ax=plt.subplots(1,2,figsize=(16,6))
color = cm.inferno(np.linspace(.4,.6, 5))
province_counts = reviews.groupby('province')['province'].count().sort_values(ascending=False)
province_counts = province_counts[:5]
province_counts.plot(kind='bar',color=color,legend='Province',ax=ax[0]).set_title('Count for province in the market of wines')
province_counts_pct = province_counts/len(reviews)
province_counts_pct.plot(kind='bar',color=color,legend='Province',ax=ax[1]).set_title('% of share in the market of wines')
plt.xlabel('');
```

```python
plt.figure(figsize=(12,6))
c2 = cm.twilight(np.linspace(.2,.6, 10))
reviews['points'].value_counts().sort_index().plot(kind='bar',color=c2).set_title('Number of awarded points for wines');
plt.xlabel('Points');
```

We can see on the chart above that there 85-90 points are most common ratings for wines.

```python
def wine(price):
    if price > 0 and price <= 20:
        return '0-20'
    elif price >20 and price <=40:
        return '21-40'
    elif price >40 and price <=80:
        return '41-80'
    elif price >80 and price <=150:
        return '81-150' 
    else:
        return '>150'
reviews['PriceBand'] = reviews['price'].apply(wine)
reviews.head(3)
```

```python
pricebands = reviews.groupby('PriceBand')['points'].count()
pricebands=pricebands.reset_index()
pricebands = pricebands.sort_values(by='PriceBand')
```

```python
pb_pct = reviews.groupby('PriceBand')['points'].count()/len(reviews)
pb_pct = pb_pct.reset_index()
```

```python
f,ax=plt.subplots(1,2,figsize=(16,6))
sns.barplot(data=pricebands,y='points',x='PriceBand',palette='cividis',ax=ax[0])
sns.barplot(data=pb_pct,y='points',x='PriceBand',palette='plasma_r',ax=ax[1]);
```

```python
from palettable.colorbrewer.qualitative import Dark2_7
f,mc = plt.subplots(figsize=(16,12))
mc = plt.Circle((0,0),0.75,color='white')
plt.pie(pb_pct['points'],colors=Dark2_7.hex_colors,autopct='%1.1f%%',startangle=30,pctdistance=0.87
        ,explode = (0.05,0.05,0.05,0.05,0.05))
p = plt.gcf()
p.gca().add_artist(mc)
plt.legend(labels=('Price Band: '  + pb_pct['PriceBand'] ),loc="upper right")
plt.show()
```

We can see on the chart above that most of wines are in under 100 USD. The biggest group almost 40% are wines under 20 USD.
So we can check histogram of prices under 100

```python
mask = reviews.groupby('variety')['points'].count()
mask = mask.sort_values(ascending=False).head(5)
#g = reviews.groupby(['variety','points'])['points']
g = reviews.groupby(['variety','points'])['points'].count()
g = pd.DataFrame(g)
g = g.rename(columns={'points':'counts'})
g = g.reset_index()
top_g = g[g['variety'].isin(mask.index.tolist())]
wine_counts = top_g.pivot(index='points',columns='variety',values='counts')
#counts_100 = top_g.pivot(index='points',columns='variety',values='counts').tail()
```

```python
c1 = cm.twilight(np.linspace(.1,.6, 5))
wine_counts.plot(kind='bar',stacked=True,color=c1).set_title('Stacked bar of points/variety',fontsize=16);
```

```python
c2 = cm.inferno(np.linspace(.1,.6, 5))
wine_counts.plot(kind='area',color=c2).set_title('Area bar of points/variety',fontsize=16);
```

```python
c3 = cm.cividis_r(np.linspace(.1,.6, 5))
wine_counts.plot(kind='line',color=c3,linewidth=4).set_title('Line plots of points/variety',fontsize=16);
```

```python
price_points = reviews[reviews['price'] < 150][['price','points']].dropna().sample(1000)
plt.figure(figsize=(14,9))
sns.set_style("white")
sns.kdeplot(price_points,cmap="Reds",cbar=True,shade=True).set_title('Points/Price KDE Plot',fontsize=14)
plt.xlabel('Price')
plt.ylabel('Points');
```

```python
sns.jointplot(x='price', y='points', data=reviews[reviews['price'] < 150], kind='hex', 
              gridsize=22,height=9);
```

```python
varieties5 = reviews.groupby('variety')['price'].agg(['mean','std','count']).sort_values(by='count',ascending=False).head(5)
top_5 = reviews[reviews['variety'].isin(varieties5.index)]
```

```python
f,ax = plt.subplots(2,1,figsize=(18,16))
sns.boxplot(
    x='variety',
    y='points',
    data=top_5,ax=ax[0]);
sns.swarmplot(x='variety', y='points', data=top_5.sample(500),ax=ax[1]);
```

_We can see on charts above that:_
    1. Pinot Noir: Most of Wines are between 86-92 points
    2. Red Blend:  Most of Wines are between 86-91 points
    3. Chardonnay:  Most of Wines are between 85-90 points
    4. Cabernet Sauvignon:  Most of Wines are between 86-91 points
    5. Sauvigon Blanc:  Most of Wines are between 85-89 points

```python
plt.subplots(figsize=(12,8))
sns.swarmplot(x='variety', y='price', data=top_5[top_5['price']<150].sort_values(by='variety').sample(500));
plt.title('Prices for most popular wines variety under 150 USD [500 Sample]',fontsize=15);
```

On the swarmplot above we can see that when we take sample of 500 wines from top 5 most popular varieties with prices under 150. We can see that for almost all of them are between 10 to 80 USD.
